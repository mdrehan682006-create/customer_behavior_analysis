```python
import pandas as pd

df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.isnull().sum()
```

```python
#create a column age_group
labels = ['Young Adult','Adult','Middle_Aged','Senior']
df['age_group'] = pd.qcut(df['Age'], q=4, labels = labels)
```

```python
df[['Age','age_group']].head(10)
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','-')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create columns purchase_frequency_days

frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually':365,
    'Every 3 Months':90
}

df['purchase_frequency_days'] = df['frequency-of-purchases'].map(frequency_mapping)

 
```

```python
df[['purchase_frequency_days','frequency-of-purchases']].head(10)
```
